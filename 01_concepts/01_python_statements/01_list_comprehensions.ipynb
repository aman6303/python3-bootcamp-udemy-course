```python
# nested loops in list comprehension (not recommended)

mylist = []

for x in [2, 4,6 ]:
    for y in [1, 20, 300]:
        mylist.append(x*y)
        
mylist
```

```python
mylist = [x*y for x in [2, 4, 6] for y in [1, 10, 100]]
mylist
```
