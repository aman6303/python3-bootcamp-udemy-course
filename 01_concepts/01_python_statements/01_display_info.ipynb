# Taking validating user input

```python
# we have to check two conditions
# if it is a digit or within a range

def user_choice():
    
    choice = "Wrong"
    accepted_range = range(10)
    acc = False
    
    while (choice.isdigit() == False) or acc == False:
        choice = input("Enter a number(0-10): ")
        
        if choice.isdigit() == False:
            print("Please enter a valid number")
        if choice.isdigit() == True:
            if int(choice) in accepted_range:
                acc = True
            else:
                print("No should be in range")
    return int(choice)
           
```

# Building a simple list game

```python
def display_mygame(game_list):
    print("Current position of List: ",game_list)
```

```python
def position_update():
    
    choice = input("Enter your postion you want to change: ")
    
    return int(choice)
```

```python
def ask_value():
    
    value = input("Enter your value you want to change: ")
    
    return value
```

```python
def update_game(value, position_value, game_list):
    game_list[position_value] = value
    return game_list
```

```python
def want_to_continue():
    
    want_to_play = True
    while want_to_play:
        want_to_play = input("Want to Continue (Y/N): ").lower()
        if want_to_play == 'y':
            return True
        elif want_to_play == 'n':
            return False
        else:
            print("Please Enter a valid ans (y/n)!!!")
```

```python

while want_to_continue():
    game_list = [0,1,2]
    display_mygame(game_list)
    value = ask_value()
    position_value = position_update()
    
    game_list = update_game(value, position_value, game_list)
    display_mygame(game_list)
else:
    print("Thanks for playing") 
```
